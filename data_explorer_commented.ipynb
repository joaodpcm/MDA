## Extract percentiles 

```python
# Defining function for extracting the percentiles on the peaks 
def column_dropper_42(df):
    df = df.drop(['description', '#object_id',
                'lamax_unit', 'lamax',
                  'laeq_unit', 'lceq',
                  'lceq_unit', 'lcpeak',
                 'lcpeak_unit'], axis=1)
    df = df.set_index('result_timestamp')
    df = df.dropna(how='all')
    df = df.rename(columns={'laeq':'laf25_per_hour'})
    df = df.resample('1H').agg({'laf25_per_hour': lambda x:resample(x)})
    df = df.dropna()
    df = df.reset_index()
    return df

def resample(array_like):
    return np.nanpercentile(array_like, 75)


filosofia_42_df_small = column_dropper_42(filosofia_42_df)
filosofia_full = pd.concat([filosofia_40_df, filosofia_42_df_small])
filosofia_full = filosofia_full.sort_values(by='result_timestamp', ascending=True)
filosofia_full = filosofia_full.reset_index()
filosofia_full = filosofia_full.drop(columns=['index'])

naam62_42_df_small = column_dropper_42(naam62_42_df)
naam62_full = pd.concat([naam62_40_df, naam62_42_df_small])
naam62_full = naam62_full.sort_values(by='result_timestamp', ascending=True)
naam62_full = naam62_full.reset_index()
naam62_full = naam62_full.drop(columns=['index'])

naam81_42_df_small = column_dropper_42(naam81_42_df)
naam81_full = pd.concat([naam81_40_df, naam81_42_df_small])
naam81_full = naam81_full.sort_values(by='result_timestamp', ascending=True)
naam81_full = naam81_full.reset_index()
naam81_full = naam81_full.drop(columns=['index'])

calvar_42_df_small = column_dropper_42(calvar_42_df)
calvar_full = pd.concat([calvar_40_df, calvar_42_df_small])
calvar_full = calvar_full.sort_values(by='result_timestamp', ascending=True)
calvar_full = calvar_full.reset_index()
calvar_full = calvar_full.drop(columns=['index'])
```

## Merge noise stations dfs

```python
# defining function for reducing the data set to the selected features
def df_reducer(df, names, i):
    df_small = df.drop(['laf005_per_hour', 'laf005_per_hour_unit',
                'laf01_per_hour', 'laf01_per_hour_unit',
                  'laf05_per_hour', 'laf05_per_hour_unit',
                  'laf10_per_hour_unit',
                 'laf50_per_hour_unit',
                 'laf75_per_hour', 'laf75_per_hour_unit',
                 'laf90_per_hour', 'laf90_per_hour_unit', 
                 'laf95_per_hour', 'laf95_per_hour_unit',
                 'laf98_per_hour', 'laf98_per_hour_unit',
                 'laf99_per_hour', 'laf99_per_hour_unit',
                 'laf995_per_hour', 'laf995_per_hour_unit',
                 'laf25_per_hour_unit', 'laf005_per_hour_unit', 'description', '#object_id'], axis=1)
#     df_small = df_small.set_index('result_timestamp')
    df_small = df_small.rename(columns={'laf25_per_hour':names[i]})
    i += 1
    return df_small, i

names = ['filosofia', 'naam35', 'naam57', 'naam62', 'naam81', 'calvar']
i = 0
naam35_full = naam35_40_df.copy()
naam57_full = naam57_40_df.copy()
filosofia_small, i = df_reducer(filosofia_full, names, i)
naam35_small, i = df_reducer(naam35_full, names, i)
naam57_small, i = df_reducer(naam57_full, names, i)
naam62_small, i = df_reducer(naam62_full, names, i)
naam81_small, i = df_reducer(naam81_full, names, i)
calvar_small, i = df_reducer(calvar_full, names, i)
data_frames = [filosofia_small, naam35_small, naam57_small, naam62_small, naam81_small, calvar_small]
df_merged = reduce(lambda  left,right: pd.merge(left,right,on=['result_timestamp'],
                                            how='outer'), data_frames)
# Merging the locations from the data set 
df_merged['mean'] = df_merged[['filosofia', 'naam35', 'naam57', 'naam62', 'naam81', 'calvar']].mean(axis=1)
df_merged['scaled'] = np.zeros(len(df_merged))
df_merged['comp'] = np.zeros(len(df_merged))
df_indexed = df_merged.set_index('result_timestamp')
for hour in range(24):
    for day in range(7):
        df_hour = df_indexed[(df_indexed.index.hour == hour) & (df_indexed.index.dayofweek == day)]
        avg = df_hour['mean'].mean()
        std = df_hour['mean'].std()
        df_indexed.loc[df_hour.index, 'scaled'] = (df_indexed.loc[df_hour.index, 'mean'] - avg) / std
        
stds_tresh = 1
df_indexed.loc[df_indexed['scaled'] > stds_tresh, 'comp'] = 'High'
df_indexed.loc[df_indexed['scaled'] < -stds_tresh, 'comp'] = 'Low'
df_indexed.loc[(df_indexed['scaled'] < stds_tresh) & (df_indexed['scaled'] > -stds_tresh), 'comp'] = 'Intermediate'
df_indexed = df_indexed.sort_index(ascending=True)
```

```python
# getting the final data set 
final_df = df_indexed.drop(columns=['filosofia', 'naam35', 'naam57', 'naam62', 'naam81', 'calvar', 'scaled'])
# final_df = final_df.drop(columns=['level_0_x', 'index_x', 'level_0_y', 'index_y', 'level_0', 'index_x', 'index_y'])
final_df = final_df.rename(columns={'mean':'noise_level'})
final_df.to_csv('noise_data_1.0_complete.csv')
```
